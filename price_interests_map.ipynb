```python
# Store latitude and longitude in locations
locations = df[["latitude", "longitude"]]

# Fill NaN values and convert to float
rating = df["price"].astype(float)

# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=rating, 
                                 dissipating=False, max_intensity=200,
                                 point_radius=.01)


# Add layer
fig.add_layer(heat_layer)

# Display figure
# fig
```

```python
interest_rate = df["# point of interest"].tolist()

# Create bank symbol layer
interest_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content=[f"Number of points of interest: {pi}" for pi in interest_rate]
)


fig = gmaps.figure()
fig.add_layer(interest_layer)

# fig
```

```python
# Create a combined map
fig = gmaps.figure(map_type="HYBRID")

fig.add_layer(heat_layer)
fig.add_layer(interest_layer)

fig
```
